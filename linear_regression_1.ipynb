```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([2, 3, 4, 5, 6, 8, 9, 10, 12, 14])
y = np.array([3, 2, 5, 3, 7, 5, 8 , 7, 9, 9])
def value(X, w, b):
    return X*w + b;
def L_MSE(y_predict, y ):
    a = (y_predict - y)**2
    return np.mean(a);
def derivative_MSE_w(X, y, y_predict): 
    a = X*(y_predict - y);
    return np.mean(a);
def derivative_MSE_b(y, y_predict): 
    a = (y_predict - y);
    return np.mean(a);

#init
w = 1
b = 1
u = 0.0001
stored = [];
MSE = [];
for i in range(0, 1000):
    y_predict = value(X, w, b)
    stored.append((w, b));
    MSE.append(L_MSE(y_predict, y))
    w = w - u*derivative_MSE_w(X, y, y_predict);
    b = b - u*derivative_MSE_b(y, y_predict);
w_b = stored[MSE.index(min(MSE))]
w, b = w_b
print(w, b, min(MSE))
y_predict = w*X + b
plt.plot(X, y_predict)

plt.scatter(X, y)

```
